# Let's learn about birds


Let's start by looking at the overall dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
birds = pd.read_csv('../../data/birds.csv')
birds.head()
```

### Pie chart of birds

```python
#order
order_count = birds.value_counts(birds['Order'].values, sort=True)
order_count
plot = order_count.plot.pie(autopct = '%1.1f%%')
#ax.pie(order_count, autopct = '%1.1f%%')
```

```python
status_count = birds.value_counts(birds['ConservationStatus'].values, sort=True)
plot = status_count.plot.pie(autopct = '%1.1f%%')
```

### Categorized Information of Trumpeter Swan

```python
birds.loc[birds['Family'] == 'Anatidae']
```

```python
#let's look at birds in the  (family of ducks, geese, swans of 45)
subset = birds.loc[birds['Family'].isin(['Anatidae'])]
```

```python
family_order_count = subset.value_counts(subset['ConservationStatus'].values, sort=True)
plot = family_order_count.plot.pie(autopct = '%1.1f%%')
plt.ylabel("")
plt.show()
```

```python
name = subset['Name']
minLength = subset['MinLength']
maxLength = subset['MaxLength']
minWingspan = subset['MinWingspan']
maxWingspan = subset['MaxWingspan']
minMass = subset['MinBodyMass']
maxMass = subset['MaxBodyMass']
```

```python
plt.barh(name, maxLength)
plt.barh(name, minLength)
plt.rcParams['figure.figsize'] = [9, 12]
plt.ylabel("Bird")
plt.xlabel("Length")
plt.show()
```

```python
plt.barh(name, maxMass)
plt.barh(name, minMass)
plt.rcParams['figure.figsize'] = [9, 12]
plt.ylabel("Bird")
plt.xlabel("Mass")
plt.show()
```

```python
plt.barh(name, maxWingspan)
plt.barh(name, minWingspan)
plt.rcParams['figure.figsize'] = [9, 12]
plt.ylabel("Bird")
plt.xlabel("Wingspan")
plt.show()
```
