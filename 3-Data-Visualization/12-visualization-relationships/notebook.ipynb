```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
honey = pd.read_csv('../../data/honey.csv')
honey.head()
```

```python
sns.relplot(x="year", y="yieldpercol", kind="line", data=honey);
```

```python
sns.relplot(
    data=honey, 
    x="year", y="numcol",
    col="state", 
    col_wrap=3,
    kind="line")
```

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=honey['year'], y=honey['prodvalue'], data=honey, 
                        label = 'Total Production Value', legend=False)
sns.despine()
plt.ylabel('Production Value')
plt.title('Honey Production Year over Year');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=honey['year'], y=honey['priceperlb'], ax=ax2, color="r", 
                         label ='Honey Value', legend=False) 
sns.despine(right=False)
plt.ylabel('Honey Value')
ax.figure.legend();
```

## Conclusions:
- Total yield of honey per colony is sharply decreasing.
- For most states, bee colony populations are steady. For some, their bee colony populations are declining. And very few have increasing bee populations.
- Even though the total production value is increasing, it is because it follows an increase in the honey value.
- This tells the story that there are less bees, bees are making less honey, but the value is going up
